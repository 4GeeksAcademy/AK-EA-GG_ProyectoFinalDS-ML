```python
def calculate_average(value):
    # If the value contains a slash
    if '/' in value:
        # Split the value by slash and convert to integers
        nums = [int(num) for num in value.split('/')]
        # Calculate and return the average
        return sum(nums) / len(nums)
    else:
        # If no slash, return the original value
        return int(value)
```

```python
merged_data['City_MPG_Removed_slash'] = merged_data['City MPG'].apply(calculate_average)
merged_data['Hwy_MPG_Removed_slash'] = merged_data['Hwy MPG'].apply(calculate_average)
merged_data['Cmb_MPG_Removed_slash'] = merged_data['Cmb MPG'].apply(calculate_average)
```
